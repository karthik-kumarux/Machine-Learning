```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Category': ['A', 'B', 'C', 'D', 'E'],
    'Values': [23, 45, 56, 78, 33]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Plotting bar chart
df.plot(kind='bar', x='Category', y='Values', title='Bar Plot', color='skyblue')

# Labeling
plt.ylabel('Values')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'Age': [21, 22, 23, 24, 25, 30, 35, 40, 45, 50]}

# Creating DataFrame
df = pd.DataFrame(data)

# Plotting histogram
df['Age'].plot(kind='hist', bins=5, title='Histogram of Age', color='lightgreen')

# Labeling
plt.xlabel('Age')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [1500, 1800, 2000, 2200, 2500]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Plotting line chart
df.plot(kind='line', x='Month', y='Sales', title='Sales Over Time', color='purple', marker='o')

# Labeling
plt.ylabel('Sales')
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {
    'Age': [21, 22, 23, 24, 25],
    'Salary': [2000, 2200, 2500, 2700, 3000]
}

# Creating DataFrame
df = pd.DataFrame(data)

# Plotting scatter plot
df.plot(kind='scatter', x='Age', y='Salary', title='Scatter Plot of Age vs Salary', color='red')

# Labeling
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
