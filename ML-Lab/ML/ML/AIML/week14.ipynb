```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

# Step 1: Create and save sample data
data = {
    'X': [1.0, 1.5, 5.0, 8.0, 1.0, 9.0, 8.0, 10.0, 9.0, 2.0],
    'Y': [2.0, 1.8, 8.0, 8.0, 0.6, 11.0, 2.0, 2.0, 3.0, 2.5]
}
df = pd.DataFrame(data)
csv_file = "kmeans_performance_data.csv"
df.to_csv(csv_file, index=False)

# Step 2: Load dataset
df = pd.read_csv(csv_file)
features = df[['X', 'Y']].values  # Fixed incorrect indexing

# Step 3: Evaluate performance for various k values
performance_results = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
    kmeans.fit(features)
    
    # Calculate sum of Euclidean distances
    closest, distances = pairwise_distances_argmin_min(features, kmeans.cluster_centers_)
    total_distance = distances.sum()
    
    performance_results.append((k, total_distance))
    print(f"k = {k} -> Sum of Euclidean Distances: {total_distance:.4f}")

# Step 4: Plot performance vs k (Elbow Method)
ks, distances = zip(*performance_results)
plt.figure(figsize=(8, 6))
plt.plot(ks, distances, marker='o')
plt.title('Performance vs. Number of Clusters (k)')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Total Euclidean Distance to Cluster Centers')
plt.grid(True)
plt.show()
```
